```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Load grayscale image
image_path = '/content/Annotation 2025-10-24 154253.png'  # Path to image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Corrected image reading

# Define parameters for contrast and brightness
alpha = 2.8  # Contrast control (1.0–3.0)
beta = -50    # Brightness control (0–100)

# Apply contrast and brightness adjustment
adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Display original and adjusted images using OpenCV
cv2_imshow(image)
cv2_imshow( adjusted)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using matplotlib
plt.figure(figsize=(10, 5))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Display contrast-adjusted image
plt.subplot(1, 2, 2)
plt.imshow(adjusted, cmap='gray')
plt.title('Contrast Adjusted Image')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
# Load grayscale image
image_path = '/content/Annotation 2025-10-24 154253.png'  # Path to image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Corrected image reading

# Calculate histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
# Display histogram
plt.figure(figsize=(6,4))
plt.plot(hist, color='black')
plt.title('Histogram of Image')
plt.xlabel('Pixel Intensity (0-255)')
plt.ylabel('Frequency')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
# Load grayscale image
image =cv2.imread('/content/download.jfif',
cv2.IMREAD_GRAYSCALE)
# Apply histogram equalization
equalized = cv2.equalizeHist(image)
# Display images and histograms
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(2,2,2)
plt.hist(image.ravel(), 256, [0,256])
plt.title('Original Histogram')
plt.subplot(2,2,3)
plt.imshow(equalized, cmap='gray')
plt.title('Equalized Image')
plt.subplot(2,2,4)
plt.hist(equalized.ravel(), 256, [0,256])
plt.title('Equalized Histogram')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Step 1: Read the noisy image
img = cv2.imread('/content/output-onlinejpgtools.jpg')
# Step 2: Apply Mean Filter with a 5x5 kernel
mean_filtered = cv2.blur(img, (5,5))
# Step 3: Display the original and filtered images
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(mean_filtered, cv2.COLOR_BGR2RGB))
plt.title("Mean Filtered (5x5)")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
# Load grayscale image
#Step 1: Read grayscale image
gray = cv2.imread('/content/Annotation 2025-10-24 154253.png', 0)
# Step 2: Add artificial salt-and-pepper noise
noisy = gray.copy()
prob = 0.02
black = np.random.rand(*gray.shape) < prob
white = np.random.rand(*gray.shape) > (1 - prob)
noisy[black] = 0
noisy[white] = 255
# Step 3: Apply Median Filter
median_filtered = cv2.medianBlur(noisy, 5)
# Step 4: Display all images
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(gray, cmap='gray')
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(noisy, cmap='gray')
plt.title('Noisy Image')
```
