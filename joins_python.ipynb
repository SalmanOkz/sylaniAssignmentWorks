```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

left_table = {'Date': ['01-01-2020', '02-01-2020', '03-01-2020', '04-01-2020' ],
              'CountryID': [1,1,3,2],
              'Units': [40, 25, 30, 35]}
```

```python
df = pd.DataFrame(left_table)
```

```python
df.head()
```

```python
right_table = {'ID': [3,4],
              'Country': ['Pandama', 'Spain']}

rt = pd.DataFrame(right_table)
rt
```

```python
inner_join = df.merge(rt, left_on='CountryID', right_on='ID', how='inner')
inner_join
```

```python
inner_join = df.merge(rt, left_on='CountryID', right_on='ID', how='inner').drop('ID',axis=1)
inner_join
```

```python
left_table = {'Date': ['01-01-2020', '02-01-2020', '03-01-2020', '04-01-2020' ],
              'CountryID': [1,1,3,2],
              'Units': [40, 25, 30, 35],
              'Country': ['USA', 'USA', 'Spain', 'UK']}
```

```python
lt = pd.DataFrame(left_table)
lt
```

```python
rt
```

```python
inner_join = lt.merge(rt, left_on='Country', right_on='Country', suffixes=['_left','_right']).drop(['ID','Units'],axis=1)
inner_join
```

```python
right_join_df = rt.merge(lt, left_on='ID', right_on='CountryID', how='right',indicator=True)
right_join_df
```
